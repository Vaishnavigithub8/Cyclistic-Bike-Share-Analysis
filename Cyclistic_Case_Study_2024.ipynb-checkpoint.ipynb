## PREPARE

### Data Source
The dataset was obtained from Divvy’s publicly available historical trip data. It contains trip-level details for Cyclistic bike rides.

### Data Description
Each row represents one bike trip and includes information such as:
- Ride ID
- Rideable type
- Start and end timestamps
- Start and end station names
- Geographic coordinates
- Rider type (member or casual)

### Data Credibility (ROCCC Check)

Reliable: Public dataset provided by the bike-share company.  
Original: Primary ride data collected directly from the bike-share system.  
Comprehensive: Includes trip-level data across 12 months.  
Current: Using full year 2025 data for recent behavioral trends.  
Cited: Publicly available data source.
